```python
import re

def clean_text(text):

    text = str(text).lower()

    # remove urls
    text = re.sub(r"http\S+", "", text)

    # remove special characters
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
final_df["text"] = final_df["text"].apply(clean_text)
```

```python
final_df = final_df[
    final_df["text"].str.strip() != ""
]
```
